```python
import numpy as np 
import pandas as pd 
import descartes
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
Med_map = gpd.read_file('/Users/williz/Desktop/ModelosED/ACO/Malla_vial/Malla_vial.shp')
```

```python
fig,ax = plt.subplots(figsize = (15,15))
Med_map.plot(ax=ax)
```

```python
df = pd.read_csv('/Users/williz/Desktop/ModelosED/ACO/NodosMed.csv', sep='\t')
crs = {'init': 'epsg:4326'}
df.head()
```

```python
geometry = [Point(xy) for xy in zip(df['longitud'],df['latitud'])]
geometry[:3]
```

```python
geo_df = gpd.GeoDataFrame(df,
                         crs = crs,
                         geometry = geometry)
geo_df.head()
```

```python
fig,ax = plt.subplots(figsize = (20,20))
Med_map.plot(ax=ax, alpha=0.4, color="grey")
geo_df.plot(ax=ax, markersize = 20, color ="blue", marker="o")
plt.legend(prop={'size':15})
```

```python
BBox = (df.longitud.min(), df.longitud.max(),
        df.latitud.min(), df.latitud.max())
print(BBox)
```

```python
Mapa_Med = plt.imread('/Users/williz/Desktop/ModelosED/ACO/MapaMedellin.png')
```

```python
fig, ax = plt.subplots(figsize = (15,15))

ax.scatter(df.longitud, df.latitud, zorder=1, alpha=0.2, c='b', s=10)

ax.set_title('Ploting Spatial Medellin')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2], BBox[3])
print(ax.set_xlim)
print(ax.set_ylim)

ax.imshow(Mapa_Med, zorder=0, extent= BBox, aspect='equal')
plt.show()
```
